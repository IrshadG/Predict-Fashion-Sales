```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from torchmetrics.regression import MeanAbsolutePercentageError
import matplotlib
import random
from random import randint
import math
from tqdm import tqdm

pd.set_option('display.max_columns', None)
plt.style.use('fivethirtyeight')

SMALL_SIZE = 8
matplotlib.rc('font', size=SMALL_SIZE)

random.seed = 21
torch.seed = 21
np.seed = 21
```

```python
class Loss_Evaluation():
    def __init__(self,training=True,device='cuda:0'):
        self.device = device
        if training:
            self.type = 'Train'
        else:
            self.type = 'Test'
        self.MAPE = MeanAbsolutePercentageError().to('cuda:0')
        self.MSE = nn.MSELoss()
        self.MAE = nn.L1Loss()
        
        self.tr_mape_losses = []
        self.tr_mae_losses = []
        self.tr_mse_losses = []
        
        self.te_mape_losses = []
        self.te_mae_losses = []
        self.te_mse_losses = []
        
    def calculate_loss(self,y_hat,y,training:bool=None):
        y = torch.tensor(y)
        y_hat = torch.tensor(y_hat)
        mse = self.MSE(y_hat,y).item()
        mae = self.MAE(y_hat,y).item()
        mape = self.MAPE(y_hat,y).item()
        
        if training is None:
            if self.type == 'Test':
                training = False
            else:
                training = True
#         print('training: ',training)
        if training:
            self.tr_mse_losses.append(mse)
            self.tr_mae_losses.append(mae)
            self.tr_mape_losses.append(mape)
        else:
            self.te_mse_losses.append(mse)
            self.te_mae_losses.append(mae)
            self.te_mape_losses.append(mape)
        return (mse,mae,mape)
        
    def print_matrics(self):
        if len(self.tr_mse_losses) != 0:
            print(f'Average Train MSE loss: {round(sum(self.tr_mse_losses)/len(self.tr_mse_losses),7)}')
            print(f'Average Train MAE loss: {round(sum(self.tr_mae_losses)/len(self.tr_mae_losses),4)}')
            print(f'Average Train MAPE loss: {round(sum(self.tr_mape_losses)/len(self.tr_mape_losses),4)}') 
        if len(self.te_mse_losses) != 0:
            print(f'Average Test MSE loss: {round(sum(self.te_mse_losses)/len(self.te_mse_losses),7)}')
            print(f'Average Test MAE loss: {round(sum(self.te_mae_losses)/len(self.te_mae_losses),4)}')
            print(f'Average Test MAPE loss: {round(sum(self.te_mape_losses)/len(self.te_mape_losses),4)}') 
    
    def get_losses(self):
        return {
            'train':{
                'mse': self.tr_mse_losses,
                'avg mse': round(sum(self.tr_mse_losses)/len(self.tr_mse_losses),7),
                'mae': self.tr_mae_losses,
                'avg mae': round(sum(self.tr_mae_losses)/len(self.tr_mae_losses),4),
                'mape': self.tr_mape_losses,
                'avg mape': round(sum(self.tr_mape_losses)/len(self.tr_mape_losses),4),
                'type': self.type,                
            },
            'test':{
                'mse': self.te_mse_losses,
                'avg mse': round(sum(self.te_mse_losses)/len(self.te_mse_losses),7),
                'mae': self.te_mae_losses,
                'avg mae': round(sum(self.te_mae_losses)/len(self.te_mae_losses),4),
                'mape': self.te_mape_losses,
                'avg mape': round(sum(self.te_mape_losses)/len(self.te_mape_losses),4),
                'type': self.type,                
            },
        }
    
    def save_evaluation(self, model_name:str, description:str,lr:float, epochs:float=0,input_size:int=0,
                        hidden_size:int=0,output_size:int=0):
        tr_mse = round(sum(self.tr_mse_losses)/len(self.tr_mse_losses),7)
        tr_mae = round(sum(self.tr_mae_losses)/len(self.tr_mae_losses),4)
        tr_mape = round(sum(self.tr_mape_losses)/len(self.tr_mape_losses),4)
        
        te_mse = round(sum(self.te_mse_losses)/len(self.te_mse_losses),7)
        te_mae = round(sum(self.te_mae_losses)/len(self.te_mae_losses),4)
        te_mape = round(sum(self.te_mape_losses)/len(self.te_mape_losses),4)
        
        idx = 0
        
        f = open(r"C:\Users\admim\Dev\MMAGI\TrendSalesPrediction\models_details.json")
        data = json.load(f)
        for key, value in data.items():
            if model_name in key:
                idx = int(key[len(model_name)+1])+1
        
        details = {
            f'{model_name}_{idx}': {
                'description': description,
                'lr': lr,
                'epochs': epochs,
                'input_size':input_size,
                'output_size':output_size,
                'hidden_size':hidden_size,
                'train mse':tr_mse,
                'train mae':tr_mae,
                'train mape':tr_mape,
                'test mse':te_mse,
                'test mae':te_mae,
                'test mape':te_mape,
                'datetime':str(datetime.now())[:-10],
            },
        }
        
        data.update(details)
        
        with open(r"C:\Users\admim\Dev\MMAGI\TrendSalesPrediction\models_details.json", "w") as outfile: 
            json.dump(data, outfile)
        print(f'{model_name}_{idx} details saved successfully.')
```

```python
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        self.relu = nn.ReLU()
        self.hidden_size = hidden_size
        # LSTM1
        self.lstm1 = nn.LSTM(input_size[1], hidden_size,num_layers=2)
        self.fc1 = nn.Linear(hidden_size, output_size)
        self.fc1_1 = nn.Linear(output_size, output_size)
        # LSTM2
        self.lstm2 = nn.LSTM(input_size[1], hidden_size,num_layers=2)
        self.fc2 = nn.Linear(hidden_size, output_size)
        self.fc2_1 = nn.Linear(output_size, output_size)
        
        self.fc_cat = nn.Linear(2*output_size, output_size)        

    def forward(self, input):
        batch_size = input.size(0)
        hidden = self.init_hidden(batch_size)
        
        lstm_out1, hidden1 = self.lstm1(input[0][0].unsqueeze(0).view(len(input), batch_size, -1).float().to('cpu'), hidden)
        output1 = self.fc1(lstm_out1[-1])
        output1_1 = self.fc1_1(output1)
        
        lstm_out2, hidden2 = self.lstm2(input[0][1].unsqueeze(0).view(len(input), batch_size, -1).float().to('cpu'), hidden)
        output2 = self.fc2(lstm_out2[-1])
        output2_1 = self.fc2_1(output2)
        
        cat = torch.cat([output1_1, output2_1],dim=1)
        output = self.fc_cat(cat)
        self.relu(output)
        return output

    def init_hidden(self, batch_size):
        return (torch.zeros(2, batch_size, self.hidden_size),
                torch.zeros(2, batch_size, self.hidden_size))
```

### Read Training set data

```python
# Load the time series data from a CSV file
df = pd.read_csv('../train_df.csv')
signals = pd.read_csv('../trends/gtrends_normalized.csv')
```

```python
n = len(df.index) #700
pop = []
for i in range(n):
    sg = []
    for j in range(len(signals.index)):
        t = 0.0
        cat = df['product_type_name'].iloc[i]
        col = df['colour_group_name'].iloc[i]
        app = df['graphical_appearance_name'].iloc[i]
        if cat in signals.columns:
            t+=signals[cat].iloc[j]
        if col in signals.columns:
            t+=signals[col].iloc[j]
        if cat in signals.columns:
            t+=signals[cat].iloc[j]
        sg.append(t)
    pop.append(sg)
```

```python
train_set_1 = []
train_set_2 = []

train_len = int(len(df.iloc[0,9:-4])*0.8)
max_len =  len(df.iloc[0,9:-4])

for i in range(n):
    tr = df.iloc[i,9:-4].values.astype(float)[:train_len]
    te = df.iloc[i,9:-4].values.astype(float)[train_len:]
    train_set_1.append(torch.stack((torch.tensor(tr),torch.tensor(pop[i][:train_len]))))
    train_set_2.append(torch.stack((torch.tensor(te),torch.tensor(pop[i][train_len:max_len]))))
    
train_set_1 = torch.stack(train_set_1,dim=0)
train_set_2 = torch.stack(train_set_2,dim=0)

# Convert the time series data to PyTorch tensors
train_set_1 = torch.tensor(train_set_1,dtype=float,device='cuda')
train_set_2 = torch.tensor(train_set_2,dtype=float,device='cuda')

train_set_1.shape,train_set_2.shape
```

### Read Validation set data

```python
val_df = pd.read_csv('../val_df.csv')

n = len(val_df.index) #300
pop = []
for i in range(n):
    sg = []
    for j in range(len(signals.index)):
        t = 0.0
        cat = val_df['product_type_name'].iloc[i]
        col = val_df['colour_group_name'].iloc[i]
        app = val_df['graphical_appearance_name'].iloc[i]
        if cat in signals.columns:
            t+=signals[cat].iloc[j]
        if col in signals.columns:
            t+=signals[col].iloc[j]
        if cat in signals.columns:
            t+=signals[cat].iloc[j]
        sg.append(t)
    pop.append(sg)
    
val_set_1 = []
val_set_2 = []

train_len = int(len(val_df.iloc[0,9:-4])*0.8)
max_len =  len(val_df.iloc[0,9:-4])

for i in range(n):
    tr = val_df.iloc[i,9:-4].values.astype(float)[:train_len]
    te = val_df.iloc[i,9:-4].values.astype(float)[train_len:]
    val_set_1.append(torch.stack((torch.tensor(tr),torch.tensor(pop[i][:train_len]))))
    val_set_2.append(torch.stack((torch.tensor(te),torch.tensor(pop[i][train_len:max_len]))))
    
val_set_1 = torch.stack(val_set_1,dim=0)
val_set_2 = torch.stack(val_set_2,dim=0)

print(f'Input data size: {val_set_1.shape}')
print(f'Output data size: {val_set_2.shape}')
```

### ---TRAINING MODEL---

```python
# Define the hyperparameters
input_size = (train_set_1.shape[1],train_set_1.shape[2]) # (2,84)
hidden_size = 512
output_size = train_set_2.shape[2] #21
learning_rate = 3e-03
num_epochs = 80

train_loader = DataLoader(train_set_1)

# Instantiate the model
model = LSTM(input_size, hidden_size, output_size)

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
le = Loss_Evaluation(training=True)

training_loss = []
validation_loss = []

# Train the model
for epoch in tqdm(range(num_epochs), colour="#2E8B57"):
    for i, inputs in enumerate(train_loader):
        optimizer.zero_grad()
        
        # Forward pass
        outputs = model(inputs)               
        loss = criterion(outputs[0].to('cuda:0'), train_set_2[i][0].float())
        le.calculate_loss(outputs[0].to('cuda:0'), train_set_2[i][0].float())
        
        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        
    with torch.no_grad():
        v_le = Loss_Evaluation()
        for j in range(len(val_set_1)):
            # Check training loss
            outputs = model(train_set_1[j].unsqueeze(0).to('cuda:0'))               
            v_le.calculate_loss(outputs[0].to('cuda:0'), train_set_2[j][0].float().to('cuda:0'),training=True)

            # Check validation loss
            outputs = model(val_set_1[j].unsqueeze(0).to('cuda:0'))
            v_le.calculate_loss(outputs[0].to('cuda:0'), val_set_2[j][0].float().to('cuda:0'),training=False)
        training_loss.append(v_le.get_losses()['train']['avg mse'])
        validation_loss.append(v_le.get_losses()['test']['avg mse'])    
```

```python
le.print_matrics()

plt.plot(training_loss,label='Training MSE Loss',linewidth=1.2)
plt.plot(validation_loss,label='Validation MSE Loss',linewidth=1.2)
plt.legend()
plt.show()
```
