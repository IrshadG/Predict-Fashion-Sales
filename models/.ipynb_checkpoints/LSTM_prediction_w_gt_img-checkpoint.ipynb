## Import

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from torchmetrics.regression import MeanAbsolutePercentageError
import matplotlib
import random
from random import randint
import math
from tqdm import tqdm

pd.set_option('display.max_columns', None)
plt.style.use('fivethirtyeight')

SMALL_SIZE = 8
matplotlib.rc('font', size=SMALL_SIZE)

random.seed = 21
torch.seed = 21
np.seed = 21

req = torch.load("req.pt")
```

```python
class Loss_Evaluation():
    def __init__(self,training=True,device='cuda:0'):
        self.device = device
        if training:
            self.type = 'Train'
        else:
            self.type = 'Test'
        self.MAPE = MeanAbsolutePercentageError().to('cuda:0')
        self.MSE = nn.MSELoss()
        self.MAE = nn.L1Loss()
        
        self.mape_losses = []
        self.mae_losses = []
        self.mse_losses = []
        
    def calculate_loss(self,y_hat,y,show=False):
        mse = self.MSE(y_hat.to(self.device),y.to(self.device)).item()
        mae = self.MAE(y_hat.to(self.device),y.to(self.device)).item()
        mape = self.MAPE(y_hat.to(self.device),y.to(self.device)).item()
        
        self.mse_losses.append(mse)
        self.mae_losses.append(mae)
        self.mape_losses.append(mape)
        
        if show:
            print(f'{self.type} MSE loss: {round(mse,4)}')
            print(f'{self.type} MAE loss: {round(mae,4)}')
            print(f'{self.type} MAPE loss: {round(mape,4)}')
        return (mse,mae,mape)
        
    def print_matrics(self):
        print(f'Average {self.type} MSE loss: {round(sum(self.mse_losses)/len(self.mse_losses),7)}')
        print(f'Average {self.type} MAE loss: {round(sum(self.mae_losses)/len(self.mae_losses),4)}')
        print(f'Average {self.type} MAPE loss: {round(sum(self.mape_losses)/len(self.mape_losses),4)}') 
    
    def get_losses(self):
        return {
            'mse': self.mse_losses,
            'mae': self.mae_losses,
            'mape': self.mape_losses,
            'type': self.type
        }
```

```python
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        self.relu = nn.ReLU()
        self.hidden_size = hidden_size
        # LSTM1
        self.lstm1 = nn.LSTM(input_size[1], hidden_size)
        self.fc1 = nn.Linear(hidden_size, output_size)
        # LSTM2
        self.lstm2 = nn.LSTM(input_size[1], hidden_size)
        self.fc2 = nn.Linear(hidden_size, output_size)
        
        self.fc_cat = nn.Linear(2*output_size, output_size)        

    def forward(self, input):
        batch_size = input.size(0)
        hidden = self.init_hidden(batch_size)
        
        lstm_out1, hidden1 = self.lstm1(input[0][1].unsqueeze(0).view(len(input), batch_size, -1).float().to('cpu'), hidden)
        output1 = self.fc1(lstm_out1[-1])
        
        lstm_out2, hidden2 = self.lstm2(input[0][0].unsqueeze(0).view(len(input), batch_size, -1).float().to('cpu'), hidden1)
        output2 = self.fc2(lstm_out2[-1])
        
        cat = torch.cat([output1, output2],dim=1)
        output = self.fc_cat(cat)
        self.relu(output)
        return output

    def init_hidden(self, batch_size):
        return (torch.zeros(1, batch_size, self.hidden_size),
                torch.zeros(1, batch_size, self.hidden_size))
```

### Read Training set data

```python
# Load the time series data from a CSV file
df = pd.read_csv('../train_df.csv')
signals = pd.read_csv('../trends/gtrends_normalized.csv')
```

```python
n = len(df.index) #700
pop = []
for i in range(n):
    sg = []
    for j in range(len(signals.index)):
        t = 0.0
        cat = df['product_type_name'].iloc[i]
        col = df['colour_group_name'].iloc[i]
        app = df['graphical_appearance_name'].iloc[i]
        if cat in signals.columns:
            t+=signals[cat].iloc[j]
        if col in signals.columns:
            t+=signals[col].iloc[j]
        if cat in signals.columns:
            t+=signals[cat].iloc[j]
        sg.append(t)
    pop.append(sg)
```

```python
train_set_1 = []
train_set_2 = []

train_len = int(len(df.iloc[0,9:-4])*0.8)
max_len =  len(df.iloc[0,9:-4])

for i in range(n):
    tr = df.iloc[i,9:-4].values.astype(float)[:train_len]
    te = df.iloc[i,9:-4].values.astype(float)[train_len:]
    train_set_1.append(torch.stack((torch.tensor(tr),torch.tensor(pop[i][:train_len]))))
    train_set_2.append(torch.stack((torch.tensor(te),torch.tensor(pop[i][train_len:max_len]))))
    
train_set_1 = torch.stack(train_set_1,dim=0)
train_set_2 = torch.stack(train_set_2,dim=0)

# Convert the time series data to PyTorch tensors
train_set_1 = torch.tensor(train_set_1,dtype=float,device='cuda')
train_set_2 = torch.tensor(train_set_2,dtype=float,device='cuda')

train_set_1.shape,train_set_2.shape
```

### Read Validation set data

```python
val_df = pd.read_csv('../val_df.csv')

n = len(val_df.index) #300
pop = []
for i in range(n):
    sg = []
    for j in range(len(signals.index)):
        t = 0.0
        cat = val_df['product_type_name'].iloc[i]
        col = val_df['colour_group_name'].iloc[i]
        app = val_df['graphical_appearance_name'].iloc[i]
        if cat in signals.columns:
            t+=signals[cat].iloc[j]
        if col in signals.columns:
            t+=signals[col].iloc[j]
        if cat in signals.columns:
            t+=signals[cat].iloc[j]
        sg.append(t)
    pop.append(sg)
    
val_set_1 = []
val_set_2 = []

train_len = int(len(val_df.iloc[0,9:-4])*0.8)
max_len =  len(val_df.iloc[0,9:-4])

for i in range(n):
    tr = val_df.iloc[i,9:-4].values.astype(float)[:train_len]
    te = val_df.iloc[i,9:-4].values.astype(float)[train_len:]
    val_set_1.append(torch.stack((torch.tensor(tr),torch.tensor(pop[i][:train_len]))))
    val_set_2.append(torch.stack((torch.tensor(te),torch.tensor(pop[i][train_len:max_len]))))
    
val_set_1 = torch.stack(val_set_1,dim=0)
val_set_2 = torch.stack(val_set_2,dim=0)

print(f'Input data size: {val_set_1.shape}')
print(f'Output data size: {val_set_2.shape}')
```

### ---TRAINING MODEL---

```python
# Define the hyperparameters
input_size = (train_set_1.shape[1],train_set_1.shape[2]) # (2,84)
hidden_size = 64
output_size = train_set_2.shape[2] #21
learning_rate = 3e-03
num_epochs = 80

train_loader = DataLoader(train_set_1)

# Instantiate the model
model = LSTM(input_size, hidden_size, output_size)

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
le = Loss_Evaluation(training=True)

training_loss = []
validation_loss = []

# Train the model
for epoch in tqdm(range(num_epochs), colour="#2E8B57"):
    for i, inputs in enumerate(train_loader):
        optimizer.zero_grad()
        
        # Forward pass
        outputs = model(inputs)               
        loss = criterion(outputs[0].to('cuda:0'), train_set_2[i][0].float())
        le.calculate_loss(outputs[0].to('cuda:0'), train_set_2[i][0].float())
        
        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        
    with torch.no_grad():
        t_le = Loss_Evaluation(training=True)
        v_le = Loss_Evaluation()
        
        for j in range(len(val_set_1)):
            # Check training loss
            outputs = model(train_set_1[j].unsqueeze(0).to('cuda:0'))               
            t_le.calculate_loss(outputs[0], train_set_2[j][0].float())

            # Check validation loss
            outputs = model(val_set_1[j].unsqueeze(0).to('cuda:0'))
            v_le.calculate_loss(outputs[0], val_set_2[j][0].float())
        
        mse_t = t_le.get_losses()['mse']
        training_loss.append(sum(mse_t)/len(mse_t))
        mse_v = v_le.get_losses()['mse']
        validation_loss.append(sum(mse_v)/len(mse_v))
```

```python
le.print_matrics()

plt.plot(training_loss,label='Training MSE Loss',linewidth=1.2)
plt.plot(validation_loss,label='Validation MSE Loss',linewidth=1.2)
plt.legend()
plt.show()
```

```python
# Set the model to evaluation mode
model.eval()

val_set_1_1 = []
for i in range(len(val_set_1)):
    if i in req:
        val_set_1_1.append(val_set_1[i])
val_set_1_1 = torch.stack(val_set_1_1)

le = Loss_Evaluation(training=False)

with torch.no_grad():
    for i in range(len(val_set_1_1)):
        outputs = model(val_set_1_1[i].unsqueeze(0).to('cuda:0'))
        for ii in range(len(outputs[0])):
            if outputs[0][ii] < 0:
                outputs[0][ii] = 0
        # Evaluate Loss
        le.calculate_loss(outputs[0], val_set_2[i][0])
        
le.print_matrics()
```
